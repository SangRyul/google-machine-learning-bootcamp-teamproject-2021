```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, power_transform

%matplotlib inline
```

```python
df_train = pd.read_csv('./train.csv')
df_test = pd.read_csv('./test.csv')
```

```python
# 데이터 결측율 비교 => 거의 차이 없음

features = df_train.columns

train_min_cnt = df_train.shape[0]
test_min_cnt = df_test.shape[0]
train_max_cnt = test_max_cnt = 0

for col in features[1:-1]:
    train_isna_cnt = df_train[col].isna().sum()
    test_isna_cnt = df_train[col].isna().sum()
    
    if train_max_cnt < train_isna_cnt: train_max_cnt = train_isna_cnt
    if test_max_cnt < test_isna_cnt: test_max_cnt = test_isna_cnt
        
    if train_min_cnt > train_isna_cnt: train_min_cnt = train_isna_cnt
    if test_min_cnt > test_isna_cnt: test_min_cnt = test_isna_cnt

        
x_val = list(features[1:-1])
y_val_train = [df_train[col].isna().sum() / df_train.shape[0] * 100 for col in x_val]
y_val_test = [df_test[col].isna().sum() / df_test.shape[0] * 100 for col in x_val]

plt.subplots(1, 1, figsize=(18, 6))
plt.ylim(0, 3)
plt.ylabel('NaN %')
plt.plot(x_val[:59], y_val_train[:59])
plt.plot(x_val[:59], y_val_test[:59])
plt.show()

plt.subplots(1, 1, figsize=(18, 6))
plt.ylim(0, 3)
plt.ylabel('NaN %')
plt.plot(x_val[59:], y_val_train[59:])
plt.plot(x_val[59:], y_val_test[59:])
plt.show()
```

#### 골고루 결측치가 존재한다.
- train과 test 결측율은 거의 유사

```python
# 이상치가 있는 feature 파악
# train, test 비교 결과 이상치가 있는 특성이 동일하다.

train_outlier = []

for col in features[1:-1]:
    q1_train, q3_train = df_train[col].quantile(.25), df_train[col].quantile(.75)
    iqr_train = q3_train - q1_train
    col_now_train = df_train[col]
    if col_now_train[(q1_train-iqr_train*1.5 > df_train[col]) | (q3_train+iqr_train*1.5 < df_train[col])].count() > 0:
        train_outlier.append(col)

print(train_outlier)
print(len(train_outlier))
```

#### 89/118 비율로 대다수의 특징이 이상치를 포함하고 있다.

## Feature 생성 및 결측치 채우기

<br>

```python
# 결측치를 채우기 전, 결측치 개수 자체를 특징으로 생성한 다음 결측치를 채운다.


df_train_without_nan = df_train.copy()
df_test_without_nan = df_test.copy()

# 결측치 개수 자체를 새로운 특징으로
df_train_without_nan['missing'] = df_train[features[1:-1]].isna().sum(axis=1)
df_test_without_nan['missing'] = df_test[features[1:-1]].isna().sum(axis=1)


# 이상치가 있는 feature 는 median으로,
# 이상치가 없는 feature 는 mean으로
for idx in range(1, 119):
    if f'f{idx}' in train_outlier:
        df_train_without_nan[f'f{idx}'] = df_train_without_nan[f'f{idx}'].fillna(df_train_without_nan[f'f{idx}'].median())
        df_test_without_nan[f'f{idx}'] = df_test_without_nan[f'f{idx}'].fillna(df_test_without_nan[f'f{idx}'].median())
    else:
        df_train_without_nan[f'f{idx}'] = df_train_without_nan[f'f{idx}'].fillna(df_train_without_nan[f'f{idx}'].mean())
        df_test_without_nan[f'f{idx}'] = df_test_without_nan[f'f{idx}'].fillna(df_test_without_nan[f'f{idx}'].mean())
        
        
# 정규화하기 전 분포 정보 자체를 특징으로
df_train_without_nan['std'] = df_train[features[1:-1]].std(axis=1)
df_test_without_nan['std'] = df_test[features[1:-1]].std(axis=1)
```

```python
%%time

# 밀도 분포 plot (참고 : https://www.kaggle.com/subinium/tps-aug-simple-eda)

# 1 ~ 110
fig, axes = plt.subplots(11, 10, figsize=(20, 22))
axes = axes.flatten()

for idx, ax in enumerate(axes):
    sns.kdeplot(data=df_train_without_nan, x=f'f{idx+1}', fill=True, ax=ax)
    sns.kdeplot(data=df_test_without_nan, x=f'f{idx+1}', fill=True, ax=ax)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_title(f'f{idx+1}', loc='right', weight='bold', fontsize=10)

fig.tight_layout()
plt.show()


# 111 ~ 118, missing, std
tmp_feature = ['f111', 'f112', 'f113', 'f114', 'f115', 'f116', 'f117', 'f118', 'missing', 'std']
fig, axes = plt.subplots(1, 10, figsize=(20, 2))
axes = axes.flatten()

for idx, ax in enumerate(axes):
    sns.kdeplot(data=df_train_without_nan, x=tmp_feature[idx], fill=True, ax=ax)
    sns.kdeplot(data=df_test_without_nan, x=tmp_feature[idx], fill=True, ax=ax)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_title(tmp_feature[idx], loc='right', weight='bold', fontsize=10)

fig.tight_layout()
plt.show()
```

## 왜도 skewness 처리

<br>

```python
# 왜도 skewness 검사 |val| < 1.5

strange = dict()
for col in features[1:-1]:
    col_skew = df_train_without_nan[col].skew()
    if abs(col_skew) >= 1.5:
        strange[col] = col_skew

for col, col_skew in strange.items():
    print(f'{col:>4} :  {col_skew:>.4f}')
```

```python
# 치우친 특성 => np.log1p(x - min(x))

# 양수 데이터만 있으면 그냥 np.log1p(x)
# 음수 데이터도 있으면 np.log1p(x - min(x)) => 데이터를 양수로 평행이동 뒤 정규화
for col, _ in strange.items():
    if min(df_train_without_nan[col]) >= 0:
        df_train_without_nan[col] = np.log1p(df_train_without_nan[col])
    else:
        df_train_without_nan[col] = np.log1p(df_train_without_nan[col] - min(df_train_without_nan[col]))
        
    if min(df_test_without_nan[col]) >= 0:
        df_test_without_nan[col] = np.log1p(df_test_without_nan[col])
    else:
        df_test_without_nan[col] = np.log1p(df_test_without_nan[col] - min(df_test_without_nan[col]))
```

```python
features = list(features[1:-1]) + ['missing', 'std'] # 새로 추가한 특성도 정규화가 필요할까?

scaler = StandardScaler()

# id, claim 제외
df_train_without_nan = scaler.fit_transform(df_train_without_nan[features])
df_test_without_nan = scaler.transform(df_test_without_nan[features])
```
